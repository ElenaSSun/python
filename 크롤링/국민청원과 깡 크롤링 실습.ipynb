```python
def asd(num):
    url = "https://www1.president.go.kr/petitions/"
    url = url + str(num)
    html = urlopen(url)
    bsObj = BeautifulSoup(html.read(), "html.parser")
    progress = bsObj.find("div",{"class":"petitionsView_progress"}).get_text()
    title= bsObj.find("h3",{"class":"petitionsView_title"}).get_text()
    counter = bsObj.find("span",{"class":"counter"}).get_text()
    counter = int(counter.replace(",", ""))
    view = bsObj.find("div",{"class":"View_write"}).get_text()
    info = bsObj.find("ul",{"class":"petitionsView_info_list"}).get_text()
    info = info.split("\n")
    category = info.split("\n")[1][4:]
    start = info.split("\n")[2][4:]
    end = info.split("\n")[3][4:]
    per = info.split("\n")[4][3:]

    return [num, progress, title, counter, view, info, category, start, end, per]
```
